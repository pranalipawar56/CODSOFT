```python
#importing pandas 
import pandas as pd
```

```python
#reading a dataset
df=pd.read_csv("E:\\Users\\GEONIX\\Documents\\Sales.csv")
```

```python
#handeling missing values
df.isnull().sum
df.dropna()
```

```python
#remove duplicates
df=df.drop_duplicates()
```
